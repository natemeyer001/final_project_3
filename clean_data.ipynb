```python
import pandas as pd
import numpy as np
from pathlib import Path
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

```python
# Import data

df = pd.read_csv(Path('spotify_dataset.csv'))
df
```

```python
df.dtypes
```

```python
# Remove rows that contain incorrect or incomplete data 

df.drop(df[df["Genre"] == "[]"].index, inplace = True)
df.drop(df[df["Popularity"] == " "].index, inplace = True)
```

```python
# Clean data to recognize numeric metrics as numbers & drop not needed columns

df["Streams"] = df["Streams"].replace({",":""}, regex=True).apply(pd.to_numeric,1)
df['Artist Followers'] = pd.to_numeric(df['Artist Followers'])
df['Popularity'] = pd.to_numeric(df['Popularity'])
df['Danceability'] = pd.to_numeric(df['Danceability'])
df['Energy'] = pd.to_numeric(df['Energy'])
df['Loudness'] = pd.to_numeric(df['Loudness'])
df['Speechiness'] = pd.to_numeric(df['Speechiness'])
df['Acousticness'] = pd.to_numeric(df['Acousticness'])
df['Liveness'] = pd.to_numeric(df['Liveness'])
df['Tempo'] = pd.to_numeric(df['Tempo'])
df['Duration (ms)'] = pd.to_numeric(df['Duration (ms)'])
df['Valence'] = pd.to_numeric(df['Valence'])
df['Release Date'] = pd.to_datetime(df["Release Date"])
df.drop(["Index","Week of Highest Charting","Weeks Charted"], axis=1, inplace=True)
```
