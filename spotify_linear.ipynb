```python
# referencing Module 17 linear regression

import pandas as pd
from pathlib import Path
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

```python
df = pd.read_csv(Path('spotify_dataset.csv'))
df
```

```python
df.dtypes
```

```python
df2 = df[:5]
```

```python
df2
```

```python
df2['Danceability'] = pd.to_numeric(df2['Danceability'])
```

```python
df2.dtypes
```

```python
# compare Danceability to Loudness

# plt.scatter(df.Danceability, df.Loudness)
# plt.xlabel('Danceability')
# plt.ylabel('Loudness')
# plt.show()
```

```python
# compare Danceability to Energy

plt.scatter(df2.Danceability, df2.Energy)
plt.xlabel('Danceability')
plt.ylabel('Energy')
plt.show()
```

```python
# use the available data to make the most accurate predictions possible
# format the data to meet the requirement sof the Scikit-learn library
# reshape(-1,1)...-1 number of rows, unspecified...1 number of columns of 1 independent variable
# x is the independent variable (years experience) on x axis
X = df2.Danceability.values.reshape(-1, 1)
```

```python
# examine first 5 rows of x
# will get a 2D NumPy array
X[:5]
```

```python
# examine shape of x - shape will be 30 samples with a ssingle feature (column)
X.shape
```

```python
# assign target variable (salary) column to y
y = df2.Energy
```

```python
# NOTE:  X is the input data and y is the output
```

```python
# create an instance of the linear regression model
# object is instantiated, created, from sklearn.linear_model's LinearRegression class (template containing algorithms)
# from LinearRegression template, model is created to analyze the data and store information
# inititate a model
model = LinearRegression()
```

```python
# fitting or training - after model inititated, will analyze data and attempt to learn patterns in the data
# use fit()
model.fit(X, y)
```

```python
# predict - predict() method used to generate predictions
# predict shape will return (30,)  of X which is 30 predictions which a straight line can be drawn
y_pred = model.predict(X)
print(y_pred.shape)
```

```python
# the model can make predictions for years experience beyond the range in the current data
# plot the predictions as a red line against the data points
plt.scatter(X, y)
plt.plot(X, y_pred, color='red')
plt.show()
```

```python
# print the slope - model.coef_
# print the y intercept - model.intercept_
print(model.coef_)
print(model.intercept_)
```
