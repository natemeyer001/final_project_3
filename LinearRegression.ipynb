# Linear Regression

```python
X = df["Popularity"].values.reshape(-1,1)
y = df["Streams"]
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=1)

model = LinearRegression()
model.fit(X_train,y_train)
y_pred = model.predict(X_test)

r2 = r2_score(y_test, y_pred)
print(f"The r^2 score for the model is {round(r2,3)}")
```

```python
# Tester for Energy vs Loudness

X = df["Energy"].values.reshape(-1,1)
y = df["Loudness"]
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=1)

model = LinearRegression()
model.fit(X_train,y_train)
y_pred = model.predict(X_test)

r2 = r2_score(y_test, y_pred)
print(f"The r^2 score for the model is {round(r2,3)}")
```

```python
# Tester for Energy vs Loudness after unpopular songs are filtered out

X = new_df["Energy"].values.reshape(-1,1)
y = new_df["Loudness"]
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=1)

model = LinearRegression()
model.fit(X_train,y_train)
y_pred = model.predict(X_test)

r2 = r2_score(y_test, y_pred)
print(f"The r^2 score for the model is {round(r2,3)}")
```
